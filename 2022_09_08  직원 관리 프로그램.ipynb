```python
empList = [] # 전역변수 - 모든 함수에 사용기 가능

# 직원 등록
def addEmployee() :
    name = input("name = ")
    pay = int(input("pay = "))
    
    emp = Employee(name, pay)
    empList.append(emp)
    print("직원등록 완료되었습니다")
    
# 모든 직원 정보 출력
def allEmployee() :
    if len(empList) != 0 :
        for emp in empList :
            print(emp.disp())
    else :
        print("등록된 직원의 저보가 존재하지 않습니다")
        
        
# 입력받은 이름에 해당하는 회원의 정보를 출력
def oneEmployee() :
    name = input("name = ")
    check = 0
    
    for emp in empList :
        if name == emp.getName() :
            print(emp.disp())
            check = 1
            
    if check == 0 :
        print("해당 이름의 직원이 존재하지 않습니다")
        
# 화면에 출력할 메뉴를 관리하는 함수
def showMenu() :
    print()
    print("\n== 직원 관리프로그램 ==")
    print("1, 직원등록")
    print("2. 직원검색(이름)")
    print("3. 전체직 원조회")
    print("4. 프로그램 종료")
```

```python
# Handler Class = Controleer
# 실제 처리를 담당하는 클래스, 프로그램 전체의 흐롬을 관리하는 클래스

class EmpManager :
    
    def __init__(self) :
        self.empList = []
        
    def addEmployee(self) :
        name = input("name = ")
        pay = int(input("pay = "))

        emp = Employee(name, pay)
        self.empList.append(emp)
        print("직원등록 완료되었습니다")
        
        
    def allEmployee(self) :
        if len(self.empList) != 0 :
            for emp in self.empList :
                print(emp.disp())
        else :
            print("등록된 직원의 저보가 존재하지 않습니다")
            
    def oneEmployee(self) :
        name = input("name = ")
        check = 0

        for emp in self.empList :
            if name == emp.getName() :
                print(emp.disp())
                check = 1

        if check == 0 :
            print("해당 이름의 직원이 존재하지 않습니다")
            
    def showMenu(self) :
        print()
        print("\n== 직원 관리프로그램 ==")
        print("1, 직원등록")
        print("2. 직원검색(이름)")
        print("3. 전체직 원조회")
        print("4. 프로그램 종료")
```

```python
def main():
    
    handle = EmpManager()
    
    while True : 
        handle.showMenu()
        menu = input("menu = ")
        
        if menu == "1" :
            handle.addEmployee()
        elif menu == "2" :
            handle.oneEmployee()
        elif menu == "3" :
            handle.allEmployee()
        elif menu == "4" :
            print("프로그램을 종료합니다")
            break
        else :
            print("메뉴의 선택이 올바르지 않습니다.")

main()
```
