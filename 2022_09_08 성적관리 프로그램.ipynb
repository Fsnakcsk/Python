```python
# 한 명의 학생에 대한 정보를 담아서 관리하는 클래스 선언 
class Student :
    def __init__(self, hakbun, name, kor, eng, math) :
        self.hakbun = hakbun
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math
        
    def getHakbun(self) :
        return self.hakbun
    
    def getName(self) :
        return self.name
    
    def update(self, name, kor, eng, math) :
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math
        print("=== 수정완료 ===")
        
    def compute(self) :
        self.tot = self.kor + self.eng + self.math
        self.avg = self.tot / 3
        
    def show(self) :
        return "%4s %5s %5s %8.2f" %(self.hakbun, self.name, self.tot, self.avg)
```

```python
class Id :

    def __init__(self) :
        self.List = []
        self.boolean = False

    def hakbunCheck(self, hakbun) :
        for num in self.List :
            if num.getHakbun() == hakbun:
                self.boolean = True
            else :
                self.boolean = False
        
    
    def addStu(self) :
        hakbun = input("학번 : ")
        self.hakbunCheck(hakbun)
        
        if self.boolean == False :
            name = input("이름 : ")
            kor = int(input("국어 :"))
            eng = int(input("영어 :"))
            math = int(input("수학 :"))

            newStu = Student(hakbun, name, kor, eng, math)
            self.List.append(newStu)
            print("======== 등록완료 ========")
            
        else :
            print("학번이 중복되었습니다")
        
    def oneStuPrint(self) :
        hakbun = input("학번 : ")
        check = 0
        print("=========================")
        print(" 학번   이름   총점   평균")
        
        for num in self.List :
            if num.getHakbun() == hakbun:  
                num.compute()
                print(num.show())
                print("=========================")
                check = 1
                
        if check == 0 :
            print("학번이 존재하지 않습니다")

            
    def scoreUpdate(self) :
        hakbun = input("학번 : ")
        check = 0
        for num in self.List :
            if num.getHakbun() == hakbun:  
                name = input("이름 : ")
                kor = int(input("국어 :"))
                eng = int(input("영어 :"))
                math = int(input("수학 :"))        
                num.update(name, kor, eng, math)
                num.compute()
                check = 1
                
        if check == 0 :
            print("학번이 존재하지 않습니다")

    def nameAllPrint(self) :
        name = input("이름 : ")
        check = 0
        print("=========================")
        print(" 학번   이름   총점   평균")
        
        for num in self.List :
            if num.getName() == name :
                print(num.show())
                check = 1
        print("=========================")
        
        if check == 0 :
            print("학번이 존재하지 않습니다")
            
    def stuAllPrint(self) :
        if len(self.List) == 0 :
            print("학생이 없습니다.")
            
        else :
            print("=========================")
            print(" 학번   이름   총점   평균")
            for stu in self.List :
                stu.compute()
                print(stu.show())
        print("=========================")
            
    def menu(self) :
        print()
        print("\n=== 학적관리 프로그램 ===")
        print("1. 학생 등록")
        print("2. 학점 검색")
        print("3. 정보 수정 ")
        print("4. 동일 이름 출력")
        print("5. 전체 출력")
        print("6. 작업 종료")
        print("=========================")
```

```python
def main() :
    handle = Id()
    
    while True :
        handle.menu()
        menu = input("menu = ")
        
        if menu == "1" :
            handle.addStu()

        elif menu == "2" :
            handle.oneStuPrint()

        elif menu == "3" :
            handle.scoreUpdate()

        elif menu == "4" :
            handle.nameAllPrint()

        elif menu == "5" :
            handle.stuAllPrint()

        elif menu == "6" :
            print("작업 종료")
            break

        else :
            print("메뉴의 선택이 올바르지 않습니다.")

main()
```
