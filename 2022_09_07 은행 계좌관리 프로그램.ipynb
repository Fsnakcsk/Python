```python
class Bank :
    
    def __init__(self, Bank_Number, name, balance) :
        self.Bank_Number = Bank_Number
        self.name = name
        self.balance = balance
        
    def get_Bank_Number(self) :
        return self.Bank_Number
    
    def get_name(self) :
        return self.name
    
    def get_balance(self) :
        return self.balance
    
    def push_balance(self, money) :
        self.balance += money
    
    def pop_balance(self, money) :
        if money > self.balance :
            print("잔액이 부족합니다")
        else :
            self.balance -= money
            
    def disp(self) :
        print("====== 전체고객 장액조회 ======")
        print("계좌번호    고객이름    예금잔액")
        print("-----------------------------")
        print(self.Bank_Number, self.name, self.balance, sep = "\t     ")
        print("-----------------------------")
        sum = 0

        for i in cust_list.keys() :
            sum += cust_list[i].get_balance()
        print("총 보유예금액 : ",sum)
        print()
        
```

```python
cust_list = {}
        
def add_cust() :
    print("== 계좌개설 ==")
    Bank_Number = input("계좌번호 = ")
    if Bank_Number in cust_list.keys() :
        print("계좌번호 중복")
    else :
        name= input("고객이름 = ")
        balance= int(input("예금금액 = "))
        print("="*30)
        print("계좌개설이 완료되었습니다\n\n")
        newCust = Bank(Bank_Number, name, balance)
        cust_list[Bank_Number] = newCust

        
def push_Money() :
    num = input("계좌번호 = ")
    if num in cust_list.keys() :
            money = int(input("입금금액"))
            cust_list[num].push_balance(money)
    else :
        print("계좌번호가 존재하지 않습니다.")

def pop_Money() :
    num = input("계좌번호 = ")
    if num in cust_list.keys() :
            money = int(input("입금금액"))
            cust_list[num].pop_balance(money)
    else :
        print("계좌번호가 존재하지 않습니다.")
        
def all_cust() :
    for i in cust_list.keys() :
        cust_list[i].disp()

        
def get_Menu() :
    print("\n== 계좌관리 프로그램 ==")
    print("1. 계좌 개설")
    print("2. 입금처리")
    print("3. 출금처리")
    print("4. 전체고객 잔액조회")
    print("5. 프로그램 종료")
    print("=====================")
    return int(input("매뉴 = "))
    
```

```python
# print("은행 계좌관리 프로그램을 시작합니다.")

while True:
    menu = get_Menu()
    
    if menu == 1:
        add_cust()
        
    elif menu == 2:
        push_Money()
        
    elif menu == 3:
        pop_Money()
    
    elif menu == 4:
        all_cust()
        
    elif menu == 5:
        break
        
    else:
        print("메뉴의 선택이 올바르지 않습니다.")
        
print("은행 계좌관리 프로그램을 종료합니다.")
```
