```python
class Cafe :
    
    def __init__(self) :
        self.table_List = list()
        self.menu_List = list()
        self.stay_List = list()
        self.name = ""
    
    def get_Name(self) :
        return self.name
    
    def set_Name(self) :
        name = input("상호명 : ")
        self.name = name
    
    def get_Table(self) :
        return self.table_List
    
    def set_Table(self) :
        n = int(input("테이블 개수 = "))
        for i in range(1, n+1) :
            self.get_Table().append([i, 0, []]) 
    
    def clear(self) :
        self.table_List = []
        self.menu_List = []
        self.stay_List = []
        self.name = ""
        
        
# 메뉴 ------------------------------------------------------------------
    
    def get_Menu(self) :
        return self.menu_List
    
    def select_Menu(self) :
        print("-"*40)
        print("번호       제품명       금액")
        for menu in self.get_Menu() :
            print("%s %12s %10s원" %(menu[0],menu[1],menu[2]))
        print("-"*40)
    
    def set_Menu(self) :
        
        print("-- < %s > --" %self.get_Name())
        print("-"*40)
        print("메뉴 추가 : 0 = 종료")
        print("-"*40)
        
        index = 0
        while True :
            menu = input("메뉴 이름 : ")
            if menu == "0" :
                self.select_Menu()
                break
            
            else :
                price = int(input("가격 설정 : "))
                index += 1
                self.get_Menu().append([index, menu, price])
                print("= 메뉴가 추가 되었습니다 =\n")
                
# 주문 --------------------------------------------------------------

    def table_Check(self) :
        for table in self.get_Table() :
            if table[1] == 0 :
                table[1] = 1
                
                domain = len(self.get_Menu()) 
                self.select_Menu()
                
                while True :
                    num = int(input("메뉴를 선택하세요 (종료하려면 0입력) : "))
                    if str(num) == "0" :
                        break
                    
                    elif num >= 1 and num <= domain :
                        
                        name = self.get_Menu()[num-1][1]
                        price = self.get_Menu()[num-1][2]
                        
                        print("%s의 " %name, end="")
                        priceNum = int(input(name + "수량 = "))
                        tot = price * priceNum
                        table[2].append([num, name, price, priceNum, tot]) # **테이블 리스트 3중에서 4중으로 변경**

                break
                       
    def check_Out(self) :
        tableNumber = int(input("결제 - 테이블 번호 = "))
        for table in self.get_Table() :
            if table[0] == tableNumber :
                self.select_Receipt(table[2])
                table[1] = 0
                table[2] = []
                

    def select_Receipt(self, table) :
        print("=============== 영수증 ===============")
        print("------------< " + self.get_Name() + " >------------")
        print("번호    제품명     단가    수량    금액")
        
        tot = 0
        for i in table:
            print("%s %9s %9d %4d %9d원" %(i[0], i[1], i[2], i[3], i[4]))
            tot += i[4]

        print("-"*40)
        print("총 결제 금액은 "+ str(tot)+"입니다.")
        

    def add_cust(self) :
        
        if self.full_Check() == True :
            
            stay_number = len(self.get_stay())
            print("자리가 없습니다. 앞에 대기중인 손님 %d명 "%stay_number)
            name = input("대기명을 입력하세요 : ")
            self.get_stay().append(name)
            
            if stay_number > 8:
                print("죄송합니다. 해당 대기인원을 받을 수 없습니다. 다음에 저희 매장을 찾아주시길 바랍니다.")
        
        else :
            if self.get_LenStay() != 0 :
                del self.get_stay()[0]
                self.table_Check()

            else :
                self.table_Check()

                
    def full_Check(self) :
        count = 0
        for i in self.get_Table() :
            if i[1] == 1 :
                count += 1
        
        if count == len(self.get_Table()) :
            return True
    
# 대기 --------------------------------------------------------------

    def get_stay(self) :
        return self.stay_List  
    
    def get_LenStay(self) :
        return len(self.get_stay())

#--------------------------------------------------------------------

    def seting(self) :

        while True :
            print()
            print("="*40)
            print("1. 가게 OPEN\n2. 테이블 조회\n3. 고객 추가\n4. 고객 결제\n5. 가게 변경\n0. 작업 종료")
            print("="*40)
            
            num = int(input("선택 = "))
            

            if num == 1 :
                self.set_Name()
                self.set_Menu()
                self.set_Table()
                
            elif num == 2 :
                print("\n>-------- 테이블 현황 --------<\n")
                for table in self.get_Table() :
                    
                    if table[1] == 1 :
                        print("   - 테이블[%d] (%s)" %(table[0],"사용중..."))
                        
                    else :
                        print("   - 테이블[%d] (%s)" %(table[0],"비어 있음..."))
                
                print("\n>-------- 대기줄 현황 --------<\n")
                
                if len(self.get_stay()) == 0 :
                    print("   - 대기 없음 ")
                
                else :
                    print("자동 알림!!\n다음 손님은", self.get_stay()[0], "입니다. 잠시만 기다려주세요.")
                    print("-"*30)
                    String = "- 대기 현황 : " 
                    domain = self.get_LenStay()
                    for index in range(domain) :
                        
                        if index == 0 :
                            String += self.get_stay()[index]
                        
                        else : 
                            String += ", " + self.get_stay()[index]
                    print(String)
                    print("-"*30)
                        
   
            elif num == 3 :
                self.add_cust()
                
            elif num == 4 :
                self.check_Out()
                
            elif num == 5 :
                self.clear()
                self.set_Name()
                self.set_Menu()
                self.set_Table()
                
            elif num == 0 :
                break
            
            else :
                print("올바르지 않는 입력값입니다")
                
a = Cafe()
a.seting()
```
