```python
# 송승우
class Coffee :
    
    #----- 제료 -------------------------------------------
    def __init__(self) :
        #          1컴피    2물   3설탕   4프림   5컵
        self.List = [3,     3,     3,     3,    3]
        
        #          블랙커피    설탕커피   프림커피    밀크커피
        self.cList= [1000,     1000,     1500,     2000]
        
        self.number = 0
        self.boolean = False
        
    def get_Coffee(self) :
        return self.List[0]
    
    def get_Water(self) :
        return self.List[1]
    
    def get_sugar(self) :
        return self.List[2]
    
    def get_prim(self) :
        return self.List[3]
    
    def get_Cup(self) :
        return self.List[4]
    
    def get_List(self) :
        return self.List
    
    def get_cList(self) :
        return self.cList
        
    def cailiao(self) :
        print("커피%d 물%d 설탕%d 프림%d 컵%d" %(self.get_Coffee(),
                                         self.get_Water(),
                                         self.get_sugar(),
                                         self.get_prim(),
                                         self.get_Cup()))
        
# ------- 커피 주문 작업 ---------------------------------------- 
    
    def b_Pop(self) :
        if self.get_List()[0] == 0 or self.get_List()[1] == 0 or self.get_List()[4] == 0 :
            print("재료 부족으로 판매 불가한 제품입니다")
        else :
            self.get_List()[0] -= 1
            self.get_List()[1] -= 1
            self.get_List()[4] -= 1
            self.number += 1
            print("주문번호 =",self.number)

    def s_Pop(self) :
        if self.get_List()[0] == 0 or self.get_List()[1] == 0 or \
        self.get_List()[2] == 0 or self.get_List()[4] == 0 :
            print("재료 부족으로 판매 불가한 제품입니다")
        else :
            self.get_List()[0] -= 1
            self.get_List()[1] -= 1
            self.get_List()[2] -= 1
            self.get_List()[4] -= 1
            
            self.number += 1
            print("주문번호 =",self.number)
            
    def p_Pop(self) :
        if self.get_List()[0] == 0 or self.get_List()[1] == 0 or \
        self.get_List()[3] == 0 or self.get_List()[4] == 0 :
            print("재료 부족으로 판매 불가한 제품입니다")
        else :
            self.get_List()[0] -= 1
            self.get_List()[1] -= 1
            self.get_List()[3] -= 1
            self.get_List()[4] -= 1
            
            self.number += 1
            print("주문번호 =",self.number)
            
    def m_Pop(self) :
        if 0 in self.get_List() :
            print("재료 부족으로 판매 불가한 제품입니다")
        else :
            for i in range(len(self.List)) :
                self.get_List()[i] -= 1
            
            self.number += 1
            print("주문번호 =",self.number)
        
    def lingQian(self, push_Money, price) :
        if push_Money >= price :
            return True
        else :
            return False
    

    def menu(self) :
        
        while True :
            print("="*30)
            print("===== 컴피 자동 판매기 동작 =====")
            print("="*30)
            print("1. 불랙커피(%d원)" %(self.get_cList()[0]))
            print("2. 설탕커피(%d원)" %(self.get_cList()[1]))
            print("3. 프림커피(%d원)" %(self.get_cList()[2]))
            print("4. 밀크커피(%d원)" %(self.get_cList()[3]))
            print("5. 주문 종료")
            print("-"*40)
            num = input("매뉴 섵택하세요 = ")

            if num == "1" :
                push_Money = int(input("투입 금액 = "))
                if self.lingQian(push_Money, self.get_cList()[0]) == True :
                    self.b_Pop()
                else :
                    print("투입 금액이 부족합니다")

            elif num == "2" :
                push_Money = int(input("투입 금액 = "))
                if self.lingQian(push_Money, self.get_cList()[1]) == True :
                    self.s_Pop()
                else :
                    print("투입 금액이 부족합니다")

            elif num == "3" :
                push_Money = int(input("투입 금액 = "))
                if self.lingQian(push_Money, self.get_cList()[2]) == True :
                    self.p_Pop()
                else :
                    print("투입 금액이 부족합니다")

            elif num == "4" :
                push_Money = int(input("투입 금액 = "))
                if self.lingQian(push_Money, self.get_cList()[3]) == True :
                    self.m_Pop()
                else :
                    print("투입 금액이 부족합니다")

            elif num == "manager" :
                self.Manager()
                
            elif num == "5" :
                break
            
            elif num == "0" :
                self.cailiao()

# -------------------------------------------------------------------------
                
    def Manager(self) :
        while True :
            print("=========== 자동판매기 관리 프로그램 ===========")
            print("[1]재료보층  [2]금액조정  [3]작업종료\n" + "="*40)
            num = input("매뉴 = ")

            if num == "1" :
                print("============== 재료보충 ==============")
                print("[1]전체 [2]커피 [3]물 [4]설탕 [5]프림 [6]컵\n" + "="*40)
                n = int(input("선택 = "))
                if n == 1 :      
                    self.all_Add()
                else :
                    self.one_Add(n-2)

            elif num == "2" : 
                print("============== 가격변경 ==============")
                print("[1]불랙커피 [2]설탕커피 [3]프림커피 [4]밀크커피\n" + "="*40)
                n = int(input("선택 = "))
                self.set_Price(n-1)

            elif num == "3" :
                print("작업이 종료되었습니다")
                break

            else : 
                print("올바르지 않는 입력입니다")
        
    def one_Add(self, num) :
        self.get_List()[num] = 3
    
    def all_Add(self) :
        for i in range(len(self.List)) :
            self.get_List()[i] = 3
            
    def set_Price(self, num) :
        price = int(input("재설정 가격 입력하세요 = "))
        self.get_cList()[num] = price
```

```python
a = Coffee()
a.menu()
```
